```python
from crewai import Agent, Crew, Task
from crewai_tools import SerperDevTool
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
import os
api_key_1 = os.environ.get("OPENAI_API_KEY")
api_key_2 = os.environ.get("SERPER_API_KEY")

os.environ["OPENAI_API_KEY"] = api_key_1
os.environ["SERPER_API_KEY"] = api_key_2
```

```python
content_creator_backstory = ""

football_researcher = Agent(
    role = "Football Researcher",
    goal = 'Find football transfer rumours on the internet.',
    backstory = 'You are a seasoned football transfer scouter that scrapes the Internet to find best transfer rumours that are gaining the buzz.',
    tools = [SerperDevTool()],
    verbose = True,
)

content_creator = Agent(
    role = 'Social Media Content Creator',
    goal = 'Create social media post ideas based off of the rumours from Instagram, Twitter, and Facebook.',
    backstory = f'An adept and smart content creator who creates interesting and baith-worthy content across social media sites. {content_creator_backstory}',
    verbose = True,
)

get_football_transfer_rumours = Task(
    description = "Browse the internet to get transfer rumours that are creating a buzz and people are talking about it.",
    expected_output = f"List of relevant transfer rumours and sources of the rumour article/post/report.",
    agent = football_researcher,
)

develop_socialmedia_posts = Task(
    description = "Based on the transfer rumour content, create social media post ideas for platforms like Facebook, Twitter, and Instagram.",
    expected_output = "A set of ideas for the different social media platforms, minimum 5.",
    verbose = True,
    agent = content_creator,
)

crew = Crew(
    agents = [football_researcher,content_creator],
    tasks = [get_football_transfer_rumours, develop_socialmedia_posts],
    planning = True,
)

crew_result = crew.kickoff()
```
